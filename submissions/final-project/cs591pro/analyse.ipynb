```python
def cleanenglish(tweets):
    res=[]
    for tweet in tweets:
        tweet=tweet.lower()
        tweet=re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', tweet)
        tweet=re.sub('@[^\s]+', 'AT_USER', tweet)
        tweet=re.sub('[\s]+', ' ', tweet)
        tweet=re.sub(r'#([^\s]+)', r'\1', tweet)
        tweet=tweet.strip('\'"')
        res.append(tweet)
    return res
```

```python
cleaned_eng=cleanenglish(eng)
cleaneng=np.array(cleaned_eng)
```
