```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
import scipy.cluster.hierarchy as hr
from scipy.spatial.distance import pdist
import sklearn.metrics as metrics
from sklearn.cluster import KMeans
```

Here I choose menu_recipes.txt to analyse

```python
#rec = pd.read_csv('allr_recipes.txt',header=None)
#epic = pd.read_csv('epic_recipes.txt', header=None)
#maps = pd.read_csv('map.txt', header=None)
memu = pd.read_csv('menu_recipes.txt', header=None)
```

When parsing data, I count all distinct ingredient as a column. So firstly, I use a set to store all ingridients and use a list to store them as the columns number. Then I create a numpy array, make all items 0. The columns are countries and ingridients. Then go through each ingridient in each item, to make the resposding position in the matrix to be 1.

```python
print len(memu)
s=set()
cols = list()
```

```python
for ii in range(0, len(memu)):
    stmp = memu[0][ii].split('\t')
    for tmp in stmp:
        if(tmp not in s):
            s.add(tmp)
            cols.append(tmp)
```

```python
data = np.zeros((len(memu), len(cols)))
for ii in range(0, len(memu)):
    stmp = memu[0][ii].split('\t')
    for tmp in stmp:
        data[ii][cols.index(tmp)]=1
```

Try to do hierarchical clustring and draw pictures of it.

```python
Z = hr.linkage(data, method='complete', metric='euclidean')
```

```python
print Z.shape
print Z[:,0]
print Z[:,1]
print Z[:,2]
print Z[:,3]
```

```python
fig = plt.figure(figsize=(10,10))
T = hr.dendrogram(Z,color_threshold=0.4, leaf_font_size=4)
fig.show()
```

```python
distances = metrics.euclidean_distances(data)
cg = sns.clustermap(distances, method="complete", figsize=(20,20), xticklabels=False)
print cg.dendrogram_col.reordered_ind
```

```python
error = np.zeros(51)
error[0] = 0;
for k in range(1,51):
    kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
    kmeans.fit_predict(data)
    error[k] = kmeans.inertia_

plt.plot(range(1,len(error)),error[1:])
plt.xlabel('Number of clusters')
plt.ylabel('Error')
```

Since the feature defined by me is 288 dimensions, so the Error is huge here. But I find decrease more slowly afer the number of culsters is larger than 11. So here, I choose 11 as the number of centers.

```python
kmeans = KMeans(init='k-means++', n_clusters=11, n_init=10)
kmeans.fit_predict(data)
labels=kmeans.labels_
print labels
```

```python
me = memu[0][labels==0]
```
